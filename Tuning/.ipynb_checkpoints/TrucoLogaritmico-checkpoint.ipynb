```python
import pandas as pd
```

### Vamos a intentar afinar la prediction para tener un mejor score en kaggle

```python
best = pd.read_csv("prediccionesFrioFrioMerge.csv")
```

```python
best.sample(3)
```

```python
preds = best['Target']
```

```python
preds.sample()
```

```python
for i in preds :
    if i >= 0.975 or i <= 0.025:
        print(i)
```

#### Los estos pueden ser considerados casos limites, 
#### donde la penalizacion es muy dura en logaritmo si nos equivocamos

```python
#### Lo que hacemos es si alguno de estos valores se encuentra por sobre lo reemplazamos.
```

```python
aux = []
for i in preds :
    if i > 0.975 :
        aux.append(0.975)
        continue
    if i < 0.025 :
        aux.append(0.025)
        continue
        
    aux.append(i)
```

```python
best['Target'] = aux
```

```python
best.to_csv('Out/TrucoLog.csv', index=False)
```
