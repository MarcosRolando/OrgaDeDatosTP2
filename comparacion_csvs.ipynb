```
import pandas as pd
```

```
complete_training = pd.read_csv("Train_TP2_Datos_2020-2C.csv")
test = pd.read_csv("Test_TP2_Datos_2020-2C.csv")
```

```
read_columns = ["Region", "Territory", "Bureaucratic_Code", "Source ", "Billing_Country", "Account_Name", "Opportunity_Name",
                "Account_Owner", "Opportunity_Owner", "Account_Type", "Opportunity_Type", "Quote_Type", "Delivery_Terms",
                "Brand", "Product_Type", "Size", "Product_Category_B", "Currency", "Last_Modified_By", "Product_Family",
                "Product_Name", "ASP_Currency", "ASP_(converted)_Currency", "Total_Amount_Currency", "Prod_Category_A"]
training = complete_training[read_columns]
test = test[read_columns]
```

```
training_values = {}
test_values = {}
diffs = {} #(key = column, val = [(value, test/train),...])
           #Guarda los valores que estan en un set pero no en el otro

for column in read_columns:
  training_values[column] = list(training[column].value_counts().index.values)
  test_values[column] = list(test[column].value_counts().index.values)
  diffs[column] = []
  for val in training_values[column]:
    if not (val in test_values[column]):
      diffs[column].append((val, "train"))
  for val in test_values[column]:
    if not (val in training_values[column]):
      diffs[column].append((val, "test"))

print(diffs)

print("")
print("Las columnas que tienen valores en test que no estan en train son:")
for key in diffs.keys():
  for val in diffs[key]:
    if val[1] == "test":
      print(key)
      break
```

```
def func(x):
  if len(x) == 1:
    return True
  i = 0
  for val in x:
    if i == 0:
      i+=1
      anterior = val
      continue
    if val != anterior:
      return False
    anterior = val
  return True
```

```
complete_training.groupby("Opportunity_ID").agg({"TRF":func})["TRF"].value_counts(normalize=True)
```
